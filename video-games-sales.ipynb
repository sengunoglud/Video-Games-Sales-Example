#### Convert year column to integer

```python
#when i tried at the first time, i used something like this: df[['x']] = df[['x']].astype(int)
#but i got this: ValueError: cannot convert float NaN to integer 
```

```python
# x contained NaN
df = df[~df['Year_of_Release'].isnull()]

```

### Add a Quality column

```python
df['Quality'] = df['Global_Sales'] * df['Critic_Score']
```

```python
df.head()
```

```python
results = df.groupby('Year_of_Release').sum()
```

```python
#results.drop(['Critic_Count', 'User_Count'], axis=1)
```

```python
results[['Quality']] = results[['Quality']].astype(int)
results[['Global_Sales']] = results[['Global_Sales']].astype(int)
results[['Critic_Score']] = results[['Critic_Score']].astype(int)
```

```python
results.drop(['Critic_Count', 'User_Count'], axis=1)
```
